```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
diamonds = pd.read_csv('diamonds.csv').reset_index(drop = True)
```

```python
diamonds['clarity'].replace({'IF': 8, 'VVS1': 7, 'VVS2' : 6, 'VS1' : 5, 'VS2' : 4, 'SI1' : 3, 'SI2' : 2, 'I1' : 1}, inplace =True)

diamonds['color'].replace({'J': 1, 'I': 2, 'H': 3, 'G': 4, 'F': 5, 'E': 6, 'D': 7}, inplace = True)

diamonds['cut'].replace({'Fair' : 1, 'Good': 2, 'Very Good':3, 'Premium': 4, 'Ideal': 5}, inplace = True)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
X = X=np.log(diamonds[['carat','cut',"clarity","color", "depth", "table"]])
y = np.log(diamonds['price'])
```

```python
model.fit(X, y)
```

```python
model.score(X, y)
```

```python
rick_diamonds = pd.read_csv('rick_diamonds.csv')
```

```python
rick_diamonds['clarity'].replace({'IF': 8, 'VVS1': 7, 'VVS2' : 6, 'VS1' : 5, 'VS2' : 4, 'SI1' : 3, 'SI2' : 2, 'I1' : 1}, inplace =True)

rick_diamonds['color'].replace({'J': 1, 'I': 2, 'H': 3, 'G': 4, 'F': 5, 'E': 6, 'D': 7}, inplace = True)

rick_diamonds['cut'].replace({'Fair' : 1, 'Good': 2, 'Very Good':3, 'Premium': 4, 'Ideal': 5}, inplace = True)
```

```python
rick_diamonds["price_predicted"]=np.exp(model.predict(np.log(rick_diamonds[['carat','cut',"clarity","color", "depth", "table"]])))
```
